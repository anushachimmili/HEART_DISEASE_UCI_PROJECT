### Categorical Variables

```python
columns=['sex','cp','fbs','restecg','exang','slope','ca','thal']
data[columns]=data[columns].apply(lambda x:x.astype('category'))
```

```python
data.describe()
```

```python
fig,ax=plt.subplots(nrows=5,ncols=1,figsize=(30,40))
for i, column in enumerate(data.select_dtypes(include=['int','float']).columns.drop('target')):
    pd.crosstab(data[column],data['target']).plot.bar(ax=ax[i%5])
    #ax[i%5].set(xlabel= column,ylabel='Frequency of disease occurence')
#     start,end=ax[i%5].get_xlim()
#     ax[i%5].xaxis.set_ticks(np.arange(start,end+1,0.5))
#     ax[i%5].xaxis.set_major_locator(plt.MaxNLocator(15))
    plt.sca(ax[i%5])
#     plt.locator_params(axis='x', nbins=15)
    plt.xticks(rotation=90,fontsize=20)
    plt.yticks(rotation=90,fontsize=20)
    plt.xlabel(column,fontsize=18)
    plt.ylabel('Frequency of disease occurence',fontsize=16)
    lg=plt.legend(title='target',fontsize=18)
    lg.get_title().set_fontsize(18)
```

#####  Higher number of disease occurences  can be observed between 41-65 years
##### Resting Blood Pressure of 120-140 mm Hg has higher incidence of heart disease

```python
fig, (ax1,ax2,ax3)=plt.subplots(nrows=1,ncols=3,figsize=(30,10))
pd.crosstab(data['cp'],data['target']).plot.bar(ax=ax1)
ax1.set(xlabel='Chest Pain Type',ylabel='Frequency of disease occurence with chest pain types')
pd.crosstab(data['cp'],data['exang']).plot.bar(ax=ax2)
ax2.set(xlabel='Chest Pain Type',ylabel='Exercise induced angina')
pd.crosstab(data['exang'],data['target']).plot.bar(ax=ax3)
ax3.set(xlabel='Exercise induced angina',ylabel='Frequency of disease occurence with exertion/stress')
```

##### Incidence of heart disease also depends on the type of chest pain experienced by the patients. Pain is classified into four categories based on the following symptoms - (1) substernal pain, (2) provoked by exertion, or (3) relieved by rest or nitroglycerin to clarify whether it is typical anginal pain (has all three characteristics - cp=0), atypical anginal pain (has two characteristics, cp=1), or nonanginal pain (has one characteristic, cp=2), or asymptomatic pain (without these three characteristics, cp=3). From fig-2, we can see the relation between pain and exercise across all the classes and is significantly high in class 0

```python
fig, (ax1,ax2,ax3)=plt.subplots(nrows=1,ncols=3,figsize=(40,5))
pd.crosstab(data['sex'],data['target']).plot.bar(ax=ax1,figsize=(5,5))
ax1.set(xlabel='Gender',ylabel='Frequency of disease occurence with Gender')
pd.crosstab(data[data['sex']==1]['age'],data[data['sex']==1]['target']).plot.bar(ax=ax2,figsize=(15,5))
ax2.set(xlabel='Ages of male',ylabel='Frequency of disease occurence with age')
start, end = ax2.get_xlim()
ax2.xaxis.set_ticks(np.arange(start,end,3))
pd.crosstab(data[data['sex']==0]['age'],data[data['sex']==0]['target']).plot.bar(ax=ax3,figsize=(15,5))
ax3.set(xlabel='Ages of female',ylabel='Frequency of disease occurence with age')
start, end = ax3.get_xlim()
ax3.xaxis.set_ticks(np.arange(start,end,3))
```

### Train Test Split

```python
x=data.drop('target',axis=1)
y=data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=41)
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

## Model Evaluation

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,y_pred))
```
